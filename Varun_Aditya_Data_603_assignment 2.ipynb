```python
from pyspark import SparkContext

# Starting SparkContext
s = SparkContext("local", "WordCount")

# Loading the first text file
f = s.textFile("Chapter_text file.txt")

# word count
c_words = f.flatMap(lambda line: line.split()) \
                   .map(lambda word: (word.lower(), 1)) \
                   .reduceByKey(lambda a, b: a + b) \
                   .collect()

for word, count in c_words:
    print(f"{word}: {count}")

# Stop SparkContext
s.stop()
```

```python
import nltk
from nltk.corpus import words
import re
from nltk.tokenize import word_tokenize
from collections import defaultdict

# Download NLTK resources if not already downloaded
nltk.download('punkt')
nltk.download('words')

def read_file(filename):
    with open('Chapter_text_2.txt', 'r') as file:
        return file.read()

def find_non_english_words(text):
    q2_words = defaultdict(int)
    eng_words = set(nltk.corpus.words.words())

    words = word_tokenize(text.lower())

    for word in words:
        word = re.sub(r'[^\w\s]', '', word)
        if word and word not in eng_words:
            q2_words[word] += 1

    return q2_words

def map_reduce(filename):
    text = read_file(filename)
    number_counts = find_non_english_words(text)
    return number_counts

if __name__ == "__main__":
    number_counts = map_reduce("Chapter_text_2.txt")

    print("Non-English Words and Their Counts:")
    for word, count in sorted(number_counts.items()):
        print(f"{word}: {count}")
```
